# Finding best model for each year  

```python
# columns = ['arima_test_rmse', 'sarima_test_rmse', 'exog_test_rmse'] 

# avg_rmse_dict = {
#     'year':[],
#     'dataframe':[],
#     'model': [],
#     'avg_rmse': []  
# }

# def find_best_rmse(df, year):
#     avg_rmse = []
#     for col in columns:
#         avg_rmse_dict['year'].append(year)
#         avg_rmse_dict['dataframe'].append(df)
#         avg_rmse_dict['model'].append(col)
#         avg_rmse_dict['avg_rmse'].append(np.mean(df[col]))
#     rmse_df = pd.DataFrame(avg_rmse_dict)
#     return rmse_df
```

```python
def find_best_model(df, year, columns):
    avg_rmse = []
    for col in columns:
        print(f'Average RMSE for {col}: {np.mean(df[col])}')
        avg_rmse.append(np.mean(df[col]))
    print(f'Lowest average rmse score for {year} data = {min(avg_rmse)}')
```

```python
columns = ['arima_test_rmse', 'sarima_test_rmse', 'exog_test_rmse'] 
```

```python
find_best_model(sarimax_metrics_2017, 2017, columns)
```

```python
find_best_model(sarimax_metrics_2018, 2018, columns)
```

```python
find_best_model(sarimax_metrics_2018_static_d, 2018, columns)
```

```python
find_best_model(sarimax_metrics_2019, 2019, columns)
```

```python
find_best_model(sarimax_metrics_2020, 2020, columns)
```

```python
# 2019, 2020 -- with extra seasonal order columns : 

columns_19_20_so = ['arima_test_rmse', 'sarima_test_rmse_alltime_so', 'exog_test_rmse_alltime_so'] 
```

```python
find_best_model(sarimax_metrics_2019_custom_so, 2019, columns_19_20_so)
```

```python
find_best_model(sarimax_metrics_2020_custom_so, 2020, columns_19_20_so)
```
